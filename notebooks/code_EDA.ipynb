## ALGERIA FOREST FIRES DATASET

```python
##Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('123.csv',header = 1)
```

```python
dataset.head()
```

## Dataset Cleaning


```python
##Missing Values
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.infer_objects()
```

```python
dataset.loc[:122,'Region']=0
dataset.loc[122:,'Region']=1
df = dataset
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
##Remove Null values
df.isnull().sum()
```

```python
##Remove values
df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df.drop(122).reset_index(drop=True)
```

```python
df.columns
```

```python
## Fix spaces in columns names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
df.head()
```

```python
df = df[df['day'] != 'day']   # removes repeated header row
df = df[df['day'] != 'Bejaia Region Dataset']
df = df[df['day'] != 'Sidi-Bel-Abbes Region Dataset']
```

```python
df.dropna()
df = df.reset_index(drop=True)
```

## Change the Required Columns as Integer Data Type

```python
df[['day','month','year','Temperature','RH','Ws']] = df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
df.columns
```

## Changing Other Columns into float datatype

```python
Objects = [features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in Objects:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.head()
```

## Save Cleaned data in other file


```python
df.to_csv('Pratice_cleaned.csv',index = False)
```

## Exploratory Data Analysis(EDA)

```python
## Remove day, month ,year
df_copy = df.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
## Encoding of categories of classes :
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
df_copy.tail()
```

## Visualization


```python
## Plot density for all Features
sns.set_style('whitegrid')
df_copy.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
## Percentage for Pie Chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
## Plotting Pie Chart
classlabels = ['Fire','Not Fire']

plt.title('Pie Chart of Classes')
plt.Figure(figsize=(13,7))
plt.pie(percentage,labels=classlabels,colors='b''r ',autopct = '%1.1f%%')
plt.show()
```

## Monthly Fire Analysis

```python
dftemp = df.loc[df['Region']==1]
plt.subplots(figsize = (13,7))
sns.set_style('whitegrid')
sns.countplot(x='month', hue = 'Classes', data=df)
plt.xlabel('Months',weight = 'bold')
plt.ylabel('Number of Fires',weight = 'bold')
plt.title('Fire Analysis of Sidi Bel Region',weight = 'bold')
```

```python
dftemp = df.loc[df['Region']==0]
plt.subplots(figsize = (13,7))
sns.set_style('whitegrid')
sns.countplot(x='month', hue = 'Classes', data=df)
plt.xlabel('Months',weight = 'bold')
plt.ylabel('Number of Fires',weight = 'bold')
plt.title('Fire Analysis of Bejaia Region',weight = 'bold')
```

It is Observed that Month like July,August,September have most Number of Forest Fires for Both Regions and Less Fires Occur in June and rest Months as Calculated from Graphs
